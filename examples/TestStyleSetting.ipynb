```python
%matplotlib widget
import pubplotlib as pplt
import matplotlib.pyplot as plt
import numpy as np

pplt.set_style('aanda')
```

```python
pplt.figure()
x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.plot(x, y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Sine Wave')
plt.show()
```

```python
pplt.figure(style='presentation')
x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.plot(x, y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Sine Wave')
plt.show()
```
